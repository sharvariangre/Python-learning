```python
#Fill missing values in age using the median.
df['age'].median()
```

```python
df_age_filled=df.copy()
```

```python
df_age_filled['age']=df_age_filled['age'].fillna(df_age_filled['age'].median())
```

```python
df_age_filled['age'].isnull().sum()
```

```python
#Fill missing values in score using the mean.
df['score'].mean()
```

```python
df_score_filled=df.copy()
```

```python
df_score_filled['score']=df_score_filled['score'].fillna(df_score_filled['score'].mean())
```

```python
df_score_filled['score'].isnull().sum()
```

```python
#Verify that no numeric column has missing values left.
df_score_filled.select_dtypes(include='number')
df_score_filled.select_dtypes(include='number').isnull().sum()
```

```python
df_clean = df.copy()
df_clean['age'] = df_clean['age'].fillna(df_clean['age'].median())
df_clean['score'] = df_clean['score'].fillna(df_clean['score'].mean())
df_clean.select_dtypes(include='number').isnull().sum()
```
