```python
import pandas as pd
```

```python
df=pd.read_csv("students.csv")
df1=pd.read_csv("marks.csv")
```

```python
df.head()
df1.head()
```

```python
#merging two datasets (left)
merged=pd.merge(df,df1,on="Student_ID", how="left")
merged.head()
```

```python
#merging two datasets (right)
merged1=pd.merge(df1,df,on="Student_ID", how="right")
merged1.head()
```

```python
#merging two datasets (inner)
merged2=pd.merge(df,df1,on="Student_ID", how="inner")
merged2.head()
```

```python
#merging two datasets (outer)
merged4=pd.merge(df,df1,on="Student_ID", how="outer")
merged4.head()
```

```python
#Filtering
df=pd.read_csv("students.csv")
```

```python
df.head()
```

```python
df1=pd.read_csv("marks.csv")
df1.head()
```

```python
#Find students with marks < 80
df1[df1["Marks"]<80]
```

```python
#Find students from CS course with attendance > 85
merged=pd.merge(df,df1,on="Student_ID",how="inner")
merged[(merged["Course"]=="CS")&(merged["Attendance"]>85)]
```

```python
#Groupby
#Average marks per course
df.groupby("Course")["Student_ID"].mean().reset_index()
```

```python
#Number of students per course
df.groupby("Course")["Student_ID"].count().reset_index()
```

```python
#Merging
#Left merge Students & marks
left_merge=pd.merge(df,df1,on="Student_ID",how="left")
left_merge.head()
```

```python
#find sutdents who have no marks data
left_merge[left_merge["Marks"].isna()]
```
