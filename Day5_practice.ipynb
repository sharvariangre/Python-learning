```python
#Day5 : Data visualiztion using Matplotlib & seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data={
    "Student":["A","B","C","D","E"],
    "Marks":[78,90,85,36,63],
    "Hours_Studied":[4,6,8,3,7]
}
df=pd.DataFrame(data)
df.head()
```

```python
#Graph 1(Bar Chart)
plt.bar(df["Student"],df["Marks"])
plt.xlabel("Student")
plt.ylabel("Marks")
plt.title("Marks of Students")
plt.show()
```

```python
#Graph 2 (Line Chart)
plt.plot(df["Student"], df["Marks"], marker='o')
plt.xlabel("Student")
plt.ylabel("Marks")
plt.title("Marks Trend")
plt.show()
```

```python
#Graph 3 (Scatterplot) Matplotlib
plt.scatter(df["Hours_Studied"],df["Marks"])
plt.xlabel("Hours_Studied")
plt.ylabel("Marks")
plt.title("Hours Studied VS Marks")
plt.show()
```

```python
#Graph 3 Scatterplot as seabo
sns.scatterplot(x="Hours_Studied", y="Marks", data=df)
plt.title("Hours Studied vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.show()
```

```python
#Graph 4 Bar Plot as seaborn
sns.barplot(x="Student", y="Marks",data=df)
plt.title("Marks of Students (Seaborn)")
plt.show()
```
