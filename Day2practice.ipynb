```python
#write a function my_min(lst) that returns the smallest number in a list
lst=[10,20,30,40,50]
def my_min(lst):
    if not lst:
        raise ValueError("Empty list")
    m=lst[0]#index
    for x in lst[1:]:
        if x<m:
           m=x
    return m
print(my_min(lst))
```

```python
#Write my_max(lst) to find the largest number.
lst=[10,20,30,40,50]
def my_max(lst):
    if not lst:
        raise ValueError("Empty list")
    m=lst[0]#index
    for x in lst[1:]:
        if x>m:
           m=x
    return m
print(my_max(lst))
```

```python
#Write a function sort_list(lst) that returns the list sorted in ascending order.
lst=[4,3,2,8,9,5,7]
def sort_list(lst):
    n=len(lst)
    for i in range(n):
        min_index=i
        for j in range(i+1,n):
            if lst[j]<lst[min_index]:
                min_index = j
        lst[i], lst[min_index]=lst[min_index],lst[i]
    return lst
print(sort_list(lst))
```

```python
#Sort a list of people by age, then name.using sorted()
friends=[
    ("Umra",30),
    ("Sejal",21)
]
sorted_friends=sorted(friends,key=lambda x:(x[1],x[0]))
print(sorted_friends)
```
